```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\spngu\OneDrive\Desktop\mysql\e_commerce_sales.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns=df.columns.str.strip().str.lower().str.replace(" ","_")
df.isnull().sum()
```

```python
# Data cleaning.
# Handling missing value in Dataset using appropriate methods.
df['region']=df['region'].fillna(df['region'].mode()[0])
df['age']=df['age'].fillna(df['age'].mean())
df['unit_price']=df['unit_price'].fillna(df['unit_price'].mean())
df['total_price']=df['total_price'].fillna(df['total_price'].mean())
df['shipping_status']=df['shipping_status'].fillna(df['shipping_status'].mode()[0])
df.isnull().sum()
```

```python
# Convert the data types of required columns in the dataset.
df['order_date']=pd.to_datetime(df['order_date'],dayfirst=True)
df['unit_price']=df['unit_price'].astype(int) 
df['quantity']=df['quantity'].astype(int) 
df['age']=df['age'].astype(int)
df['total_price']=df['total_price'].astype(int)
df.info()
```

```python
# Replace categorical values in the dataset with meaningful lables.
df['gender']=df['gender'].replace({'f':'Female','F':'Female','fe':'Female','M':'Male','m':'Male','ma':'Male'})
df['region']=df['region'].replace({'N':'North','S':'South','E':'East','W':'West'})
df['product_name']=df['product_name'].replace({'Head':'Headphone','Key':'Keyboard','Lap':'Laptop','Mo':'Monitor'})
df['category']=df['category'].replace({'Access':'Accessories','Electro':'Electronics','W':'Wearables'})
df['quantity']=df['quantity'].replace({50:5,1.7:1,4.2:4})
df['shipping_status']=df['shipping_status'].replace({'d':'Delivered','In Tra':'In Transit','In Tr':'In Transit','Re':'Returned'})
df.head()
```

```python
# Create New Columns:
# Q1:Calculate total sales for each record in the dataset.
df['total_sales']=df['unit_price']*df['quantity']
# Q2:Calculate profit for each record in the dataset.
df['profit']=df['total_sales']*0.20
df.head()
```

```python
# Data Filtering.
# Q1:Filter all orders that belong to the Electronics Category.
electronics_sales=df[df['category']=='Electronics']
electronics_sales.head()
```

```python
# Q2:Filter all high_value orders where total sales are greater than 5000.
high_value_orders=df[df['total_sales']>5000]
high_value_orders.head()
```

```python
# Q3:Calculate the total sales for each region using a group-by operation.
region_sales=df.groupby('region')['total_sales'].sum()
region_sales.head()
```

```python
# Q4:Calculate the total profit for each gender using a group-by operation.
gender_sales=df.groupby('gender')['profit'].sum()
gender_sales.head()
```

```python
df.to_csv(r"e_commerce_sales.csv",index=False)
```
